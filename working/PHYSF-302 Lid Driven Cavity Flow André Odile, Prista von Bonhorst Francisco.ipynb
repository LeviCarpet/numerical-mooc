# Lid-driven cavity flow

Le but de cette manipulation est de déterminer l'état d'équilibre d'un fluide dans une cavité carré de taille $(nx * ny)$ dont le couvercle avance avec une vitesse constante $u=1$. On considère qu'il n'y a pas de fluide qui peut s'échapper de la cavité même si le couvercle se mouvoit. On suppose aussi que le fluide a un nombre de Reynolds proche de zéro.

On commence avec l'équation de Navier-Stokes pour des fluides incompressibles:

\begin{equation}
\frac{\partial u}{\partial t} + u \cdot \nabla u = -\frac{1}{\rho}\nabla p + \nu\nabla^2 u
\end{equation}

Si on met sous forme saclaire l'équation (1) pour qu'elle soit sans dimensions, on peut la réécrire comme

\begin{equation}
Re \left(\frac{\partial u^*}{\partial t} + u^* \cdot \nabla u^* \right) = -\nabla p^* + \nabla^2 u^*
\end{equation}

où $u^*$ et $p^*$ sont la vitesse et la pression sans dimension, respectivement.

Pour obtenir un écoulement de Stokes, on suppose que le nombre de Reynolds est proche de zéro. En appliquant cette supposition à l'équation (2) et en laissant tomber les stars, on obtient:
\begin{equation}
0 = - \nabla p + \nabla^2 u
\end{equation}

Ce qui simplie les choses! A présent, on applique l'opérateur rotationel de chaque côté de l'équation:

\begin{equation}
\nabla \times 0 = \nabla \times \left( - \nabla p + \nabla^2 u\right)
\end{equation}

Le terme de gauche vaut zéro tandis que le premier terme du terme de droite vaut: 
\begin{equation}
\nabla \times - \nabla p = 0
\end{equation}

car $\nabla \times \nabla \phi = 0$ avec $\phi$ un scalaire.

Finalement,

\begin{equation}
\nabla \times \nabla^2 u =\nabla^2\omega
\end{equation}

avec $\nabla \times u = \omega$ est la vorticité.  

En combinant toutes ces équations, on obtient l'équation de transport simplifiée de la vorticité pour un écoulement de Stokes: 

\begin{equation}
\nabla ^2 \omega = 0
\end{equation}

C'est une équation de Laplace.




On suppose que l'écoulement est bidimensionnel (2D) et que 2 composantes de la vitesse sont non-nulles $(u, v, w=0)$. Comme l'écoulement du fluide est incompressible, on peut définion la fonction de flot $\psi$ comme

\begin{equation}
u = \frac{\partial \psi}{\partial y} \text{   and   } v = - \frac{\partial \psi}{\partial x}
\end{equation}

En deux dimensions, la vorticité est orientée selon l'axe $z$ est est donnée par,

\begin{equation}
\omega = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y}
\end{equation}

ce qui, combiné avec l'équation précédente, nous amène à une autre équation dont la forme nous est familière:

\begin{equation}
\nabla^2 \psi = -\omega
\end{equation}

On a un système de deux équations couplées qui peut décrire l'écoulement du fluide qui a un nombre de Reynolds très petit dans une cavité dont le couvercle se mouvoit.

\begin{equation}
\nabla^2 \omega = 0
\end{equation}

\begin{equation}
\nabla^2 \psi = -\omega
\end{equation}

Si on substitue ces deux équations, on arrive à une nouvelle équation: l'équation biharmonique: 

\begin{equation}
\nabla^4 \psi= 0
\end{equation}

Avant de résoudre cette équation, nous devons respecter certaines conditions aux bords:

$$\left.\Psi \right|_{x=0}=\left.\Psi \right|_{x=nx}=\left.\Psi \right|_{y=0}=\left.\Psi\right|_{y=ny}=0$$

\begin{equation}\left.\frac{\partial \Psi}{\partial x}\right|_{x=0} = 0 \end{equation}

\begin{equation}\left.\frac{\partial \Psi }{\partial x}\right|_{x=nx} = 0 \end{equation}

\begin{equation}\left.\frac{\partial \Psi}{\partial y}\right|_{y=0} =0 \end{equation}

\begin{equation}\left.\frac{\partial \Psi}{\partial x}\right|_{y=ny} =1 \end{equation}


Pour les conditions de Dirichlet, il est clair qu'on ne veut pas de flôt en dehors de la boite. Quand au conditions de Neumann, il semble plutôt clair que puisque les bords gauche,droit et inférieur sont immobiles, le vitesse du flôt sur ces bords soit nulle, alors que pour le bord supérieur, il faut que le flôt ait une vitesse constante puique la plaque se mouvoit elle aussi à vitesse constante.



## Discrétisation

A présent, nous devons résoudre le problème. Pour cela, nous devons tout d'abord discrétiser l'équation biharmonique et nous avons choisit d'effectuer une discrétisation d'ordre 2.

On peut réécrire:

$$\nabla^4 \psi= \frac{\partial^4 \Psi}{\partial x^4}+\frac{\partial^4 \Psi}{\partial y^4}+2\frac{\partial^4 \Psi}{\partial x^2\partial y^2}$$

A présent on doit discrétiser chacun des 3 termes. 

$$\frac{\partial^4 \Psi}{\partial x^4} = \frac{\partial^2}{\partial x^2}\left(\frac{\partial^2 \Psi}{\partial x^2}\right)=\frac{\left(\frac{\partial^2 \Psi}{\partial x^2}\right)_{i+1,j}^k-2 \left(\frac{\partial^2\Psi}{\partial x^2}\right)_{i,j}^k +\left(\frac{\partial^2 \Psi}{\partial x^2}\right)_{i-1,j}^k}{\Delta x^2}$$

$$=\frac{\left(\Psi_{i+1,j}^k-2\Psi_{i,j}^k+\Psi_{i-1,j}^k\right)_{i+1,j}^k -2\left(\Psi_{i+1,j}^k-2\Psi_{i,j}^k+\Psi_{i-1,j}^k\right)_{i,j}^k+\left(\Psi_{i+1,j}^k-2\Psi_{i,j}^k+\Psi_{i-1,j}^k\right)_{i-1,j}^k}{\Delta x^4}$$

$$=\frac{\Psi_{i+2,j}^k - 2\Psi_{i+1,j}^k +\Psi_{i,j}^k -2 \Psi_{i+1,j}^k +4 \Psi_{i,j}^k -2\Psi_{i-1,j}^k + \Psi_{i,j}^k -2 \Psi_{i-1,j}^k +\Psi_{i-2,j}^k}{\Delta x^4}= \frac{\Psi_{i+2,j}^k-4\Psi_{i+1,j}^k+6\Psi_{i,j}^k-4\Psi_{i-1,j}^k+\Psi_{i-2,j}^k}{\Delta x^4}$$

De la même manière pour l'autre dérivée quatrième, nous avons:

$$\frac{\partial^4 \Psi}{\partial y^4} = \frac{\Psi_{i,j+2}^k-4\Psi_{i,j+1}^k+6\Psi_{i,j}^k-4\Psi_{i,j-1}^k+\Psi_{i,j-2}^k}{\Delta y^4}$$

Pour calculer le troisième terme:

$$2 \frac{\partial^2}{\partial x^2}\left(\frac{\partial^2 \Psi}{\partial y^2}\right) =2\left(\frac{\left(\Psi_{i,j+1}^k-2\Psi_{i,j}^k+\Psi_{i,j-1}^k\right)_{i+1,j}^k -2\left(\Psi_{i,j+1}^k-2\Psi_{i,j}^k+\Psi_{i,j-1}^k\right)_{i,j}^k+\left(\Psi_{i,j+1}^k-2\Psi_{i,j}^k+\Psi_{i,j-1}^k\right)_{i-1,j}^k}{\Delta x^2\Delta y^2}\right)$$

$$= 2\left(\frac{\Psi_{i+1,j+1}^k -2\Psi_{i+1,j}^k+\Psi_{i+1,j-1}^k-2\Psi_{i,j+1}^k+4\Psi_{i,j}^k-2\Psi_{i,j-1}^k+2\Psi_{i-1,j+1}^k-2\Psi_{i-1,j}^k+\Psi_{i-1,j-1}^k}{\Delta x^2\Delta y^2}\right)$$

En prenant $\Delta x = \Delta y$, on obtient:

$$\nabla^4 \psi = \frac{1}{\Delta x^4}\left(\Psi_{i+2,j}^k + \Psi_{i,j+2}^k + \Psi_{i-2,j}^k+ \Psi_{i,j-2}^k +2\Psi_{i+1,j+1}^k +2\Psi_{i+1,j-1}^k +2\Psi_{i-1,j+1}^k +2\Psi_{i-1,j-1}^k -8\Psi_{i+1,j}^k - 8\Psi_{i,j+1}^k -8 \Psi_{i-1,j}^k -8 \Psi_{i,j-1}^k +20 \Psi_{i,j}^k\right)$$



A présent, on va discrétiser les 4 conditions de Neumann aux bords. On va utiliser la méthode "Forward and backward finite difference" et on va s'aider de ce lien https://en.wikipedia.org/wiki/Finite_difference_coefficient pour y arriver. On peut donc écrire:

$$\left.\frac{\partial \Psi}{\partial x}\right|_{x=0} = \frac{-3/2\Psi_{i,j}^k +2 \Psi_{i+1,j}^k -1/2 \Psi_{i+2,j}^k}{\Delta x}=0$$

$$\left.\frac{\partial \Psi}{\partial x}\right|_{x=nx} = \frac{3/2\Psi_{i,j}^k -2 \Psi_{i-1,j}^k +1/2 \Psi_{i-2,j}^k}{\Delta x}=0$$

$$\left.\frac{\partial \Psi}{\partial x}\right|_{y=0} = \frac{-3/2\Psi_{i,j}^k +2 \Psi_{i,j+1}^k -1/2 \Psi_{i,j+2}^k}{\Delta y}=0$$

$$\left.\frac{\partial \Psi}{\partial x}\right|_{y=ny} = \frac{3/2\Psi_{i,j}^k -2 \Psi_{i,j-1}^k + 1/2 \Psi_{i,j-2}^k}{\Delta y}=1$$

Or, comme 
$$\left.\Psi_{i,j} \right|_{x=0}=\left.\Psi_{i,j}\right|_{x=nx}=\left.\Psi_{i,j} \right|_{y=0}=\left.\Psi_{i,j}\right|_{y=ny}=0$$

Alors,

$$\Psi_{i+1,j}^k=\frac{1}{4}\Psi_{i+2,j}^k \mbox{ pour x=0}, \text{ (a) }$$ 
$$\Psi_{i-1,j}^k=\frac{1}{4}\Psi_{i-2,j}^k \mbox{ pour x=nx}, \text{ (b) }$$
$$\Psi_{i,j+1}^k=\frac{1}{4}\Psi_{i,j+2}^k \mbox{ pour y=0},  \text{ (c) }$$
$$\Psi_{i,j-1}^k=\frac{-\Delta y}{2}+\frac{1}{4} \Psi_{i,j-2}^k \mbox{ pour y=ny}, \text{ (d) }$$

Commençons par fixer le nombre de points $nx,ny$ de notre grille, ainsi que la longueur de nos bords. Nous définissons notre point d'arrêt "l1_target_1" pour les itérations de la méthode du gradient conjugué .

```python
# Paramètres
nx = 41
ny = 41
xmin = 0.
xmax = 1.
ymin = 0.
ymax = 1.

#Target
l1_target_1 = 1e-8

# Discrétisation
dx = (xmax-xmin)/(nx-1)
dy = (ymax-ymin)/(ny-1)

# Mesh
x  = numpy.linspace(xmin,xmax,nx)
y  = numpy.linspace(ymin,ymax,ny)
X,Y = numpy.meshgrid(x,y)

# Longueur de la boîte
L = xmax-xmin    
```

Commençons par initialiser notre matrice $p_i$ initiale. Nous devons appliquer les conditions aux bords sur cette matrice. 

```python
# Initialization
p_i  = numpy.zeros((ny,nx))

p_i[0,:]=p_i[-1,:]=p_i[:,0]=p_i[:,-1]=0 # Condition de dirichlet aux bord


p_i[-2,2:-2]=-dy/2+(1/4)*p_i[-3,2:-2] # Conditions de Neumann au bord , pour y = ny(d)
p_i[1,2:-2]=(1/4)*p_i[2,2:-2] # Conditions de Neumann au bord , y = 0 (c)
p_i[1:-1,1]=(1/4)*p_i[1:-1,2] # Conditions de Neumann au bord , x=0 (a)
p_i[1:-1,-2]=p_i[1:-1,-3]*(1/4)  # Conditions de Neumann au bord , x = nx (b)
```

Remarquons que l'indexation est inversée pour l'axe des ordonnées. C'est à dire qu'en réalité on impose, par exemple, la condition aux bord y = 1 de Neumann  sur les derniers éléments de la matrice, alors qu'on parle du bord du haut.

Pour connaitre la convergence de nos solutions trouvées avec cette méthode vers la solution exacte, on va utiliser la norme L1, définie par:
$$\text{L1_norm} =\sum_{k=0}^N \left|p_k - p\right|$$

Définissons une fonction qui calcule cette norme, en prenant deux arguments "new" et "old" et qui fait leur soustracion en valeur absolue.

```python
def L1norm(new, old):
    norm = numpy.sum(numpy.abs(new-old))
    return norm
```

Nous allons maintenant parler de la méthode du gradient conjugué. Avant de l'implémenter, rappelons la méthode rapidement. 

L'équation à résoudre est du type: 
\begin{equation}
A{\bf p}={\bf 0},
\end{equation}

où A est une matrice $N \times N$. La méthode du gradient conjugué consiste à prendre une matrice initiale $\textbf{p}^0$, et à la modifier afin d'approcher la solution. Ceci est équivalent à prendre la valeurs de la grille $\textit{p}$ et à faire un pas de taille $\alpha$ dans la direction $\textbf{d}$, afin d'avoir:
\begin{equation}
{\bf p}^{k+1}={\bf p}^k + \alpha {\bf d}^k
\end{equation}

Ainsi si on choisit le bon $\alpha$ et le bon vecteur $\textbf{d^k}$ on peut converger très rapidement vers la solution. Cependant notre grille de départ $\textbf{p}^0$ n'est pas forcément solution de l'équation de Laplace. Nous devons donc dire plutôt que:

\begin{equation}
{\bf r}^k+A{\bf p}^k=0
\end{equation}

et nous allons essayer de faire tendre le terme $\textbf{r}^k$ ,nommé "résiduel", vers zéro le plus rapidement possible, afin d'avoir une solution au problème de Laplace.Nous avons donc que:
\begin{equation}
\textbf{r}^0 = -A\textbf{p}^0
\end{equation}

La méthode du gradient conjugué consiste à prendre des étapes de magnitude 
$$
\begin{equation}
\alpha = \frac{{\bf r}^k \cdot {\bf r}^k}{A{\bf d}^k \cdot {\bf d}^k}
\end{equation}
$$
et des vecteurs de direction:
$$
\begin{equation}
{\bf d}^{k+1}={\bf r}^{k+1}+\beta{\bf d}^{k}
\end{equation}
$$

où $\beta$ est donné par $\beta = \frac{{\bf r}^{k+1} \cdot {\bf r}^{k+1}}{{\bf r}^k \cdot {\bf r}^k}$.

Nous allons donc changer les valeurs de p selon:

\begin{equation}
{\bf p}^{k+1}={\bf p}^k + \alpha {\bf d}^k
\end{equation}

Ceci implique que les conditions de bords doivent non seulement être implémentées pour $\textbf{p}^k$ mais aussi pour $\textbf{d}^k$.

Pour la première itération nous fixons $\textbf{d}^0 = \textbf{r}^0$ avec les conditions aux bords implémentées sur $\textbf{d}^0$.

Ainsi la méthode se résume à faire les étapes suivantes:
Calculer ${\bf d}^0 = {\bf r}^0$ une seule fois, ensuite

1. Calculer $\alpha = \frac{{\bf r}^k \cdot {\bf r}^k}{A{\bf d}^k \cdot {\bf d}^k}$
2. Calculer ${\bf p}^{k+1}$
3. Calculer ${\bf r}^{k+1} = {\bf r}^k - \alpha A {\bf d}^k$ 
4. Calculer $\beta = \frac{{\bf r}^{k+1} \cdot {\bf r}^{k+1}}{{\bf r}^k \cdot {\bf r}^k}$
5. Calculer ${\bf d}^{k+1}={\bf r}^{k+1}+\beta{\bf d}^{k}$
6. Répéter!

Nous allons donc définir notre fonction "gradient_conj" qui prend comme argument une matrice initiale p, la discrétisation selon dx (car on suppose dx=dy pour tout ce travail) et une condition d'arret "l1_target" et qui renvoie une matrice "p" qui est solution de l'équation de Laplace une fois la condition verifiée, ainsi qu'une liste contenant toutes les normes calculées lors de la boucle, pour dresser un graphique de celle-ci en fonction du nombre d'itérations.

```python
def gradient_conj(p,dx,l1_target):
    '''
    Calcule la solution à l'équation biharmonique dans une cavité dont l'un des 
    bords se mouvoit à vitesse constante, par la méthode du gradient conjugué.
    Assume les conditions aux bords de Neumann, mais aussi qu'il n'y a pas de flôt
    en dehors de la cavité.
    
    Parameters:
    ----------
    p : Matrice 2D de flottants
        Matrice initiale
    dx: flottant
        Discrétisation selon x 
    l1_target: float
        critère d'intérruption d'itérations
        
    Returns:
    -------
    p: Matrice 2D de flottants
         Résultat de l'équation biharmonique.
    '''
    
    nx, ny = p.shape
    r = numpy.zeros((nx,ny))
    Ad = numpy.zeros((ny,nx))
    
    l1_norm=1 #On fixe une norme aléatoire d'abord.
    iterations = 0
    l1_conv=[]
    
    #Calcul du premier résiduel.
    r[2:-2,2:-2]=(-p[4: ,2:-2]-p[2:-2,4: ]-p[:-4 ,2:-2]-p[2:-2,:-4 ]-2*p[3:-1,3:-1]\
                 -2*p[3:-1,1:-3]-2*p[1:-3,3:-1]-2*p[1:-3,1:-3] +8*p[3:-1,2:-2]\
                 +8*p[2:-2,3:-1]+8*p[1:-3,2:-2]+8*p[2:-2,1:-3] -20*p[2:-2,2:-2])/dx**4

    d = r.copy()
    
    #Calcul du numérateur de alpha
    rho = numpy.sum(r*r)
    
    #Conditions aux bords de Dirichlet implémentées sur d.
    d[0,:]=d[-1,:]=d[:,0]=d[:,-1]=0
    
    #Conditions aux bords de Neumann implémentées sur d.
    d[-2,2:-2]=-dy/2+(1/4)*d[-3,2:-2]
    d[1,2:-2]=(1/4)*d[2,2:-2]
    d[1:-1,1]=(1/4)*d[1:-1,2]
    d[1:-1,-2]=d[1:-1,-3]*(1/4)

    #Calcul de Ad.
    Ad[2:-2,2:-2]=-(-d[4:,2:-2]-d[2:-2,4:]-d[:-4,2:-2]-d[2:-2,:-4]-2*d[3:-1,3:-1]\
                 -2*d[3:-1,1:-3]-2*d[1:-3,3:-1]-2*d[1:-3,1:-3] +8*d[3:-1,2:-2]\
                 +8*d[2:-2,3:-1]+8*d[1:-3,2:-2]+8*d[2:-2,1:-3] -20*d[2:-2,2:-2])/dx**4
    
    #Calcul du dénominateur de alpha
    sigma = numpy.sum(d*Ad)
    #Itérations
    while l1_norm > l1_target:
        
        pk = p.copy()
        rk = r.copy()
        dk = d.copy()
        
        alpha=rho/sigma    #Calcul de alpha : Step 1
        
        p = pk + alpha*dk  #Calcul de p : Step 2

        #Conditions aux bord sur p.
        p[0,:]=p[-1,:]=p[:,0]=p[:,-1]=0    #Conditions aux bord de Dirichlet

        p[-2,2:-2]=-dy/2+(1/4)*p[-3,2:-2] #Conditions aux bords de Neumann
        p[1,2:-2]=(1/4)*p[2,2:-2]         #Conditions aux bords de Neumann
        p[1:-1,1]=(1/4)*p[1:-1,2]         #Conditions aux bords de Neumann
        p[1:-1,-2]=p[1:-1,-3]*(1/4)       #Conditions aux bords de Neumann
        
        r = rk- alpha*Ad   # Calcul de r : Step 3

        rhop1 = numpy.sum(r*r) # Calcul du numérateur de Beta
        beta = rhop1 / rho   # Calcul de Beta : Step 4 
        rho = rhop1 # On rédéfinit rho pour la prochain calcul de alpha
    
     
        d = r + beta*dk     # Calcul de d : Step 5
         
            
        #On impose encore une fois les conditions aux bords sur d
        d[0,:]=d[-1,:]=d[:,0]=d[:,-1]=0  #Conditions aux bord de Dirichlet
    
        d[-2,2:-2]=-dy/2+(1/4)*d[-3,2:-2] #Conditions aux bords de Neumann
        d[1,2:-2]=(1/4)*d[2,2:-2]         #Conditions aux bords de Neumann
        d[1:-1,1]=(1/4)*d[1:-1,2]         #Conditions aux bords de Neumann
        d[1:-1,-2]=d[1:-1,-3]*(1/4)       #Conditions aux bords de Neumann

        #Calcul du nouveau Ad
        Ad[2:-2,2:-2]=- (-d[4:,2:-2]-d[2:-2,4:]-d[:-4,2:-2]-d[2:-2,:-4]-2*d[3:-1,3:-1]\
                 -2*d[3:-1,1:-3]-2*d[1:-3,3:-1]-2*d[1:-3,1:-3] +8*d[3:-1,2:-2]\
                 +8*d[2:-2,3:-1]+8*d[1:-3,2:-2]+8*d[2:-2,1:-3] -20*d[2:-2,2:-2])/dx**4
        
        sigma = numpy.sum(d*Ad) #Calcul du nouveau dénominateur de alpha
        
        l1_norm = L1norm(p,pk)  #On calcule la norme L1 et on arrête la boucle lorsqu'elle est plus petite que le target.
        iterations +=1          #On veut compter le nombre d'itérations de la boucle
        l1_conv.append(l1_norm) #Pour faire un graphique de la convergence
      
    print('Nombre d itérations par la méthode du gradient conjugué: {0:d}'.format(iterations))
    return p, l1_conv
```

Calculons donc notre solution avec la condition "l1_target_1 = 1e-8"  définie ci-dessus.

```python
p_cg_1, l1_conv_cg_1=gradient_conj(p_i.copy(), dx, l1_target_1)
```

Nous avons voulu comparer cette méthode avec la méthode de plus forte pente. Pour cela nous allons brièvement rappeler en quoi consiste cette méthode, mais cette fois-ci nous allons fixer une condition moins précise, afin de pouvoir comparer rapidement les résultats.

La méthode de la plus forte pente est tout à fait similaire à la méthode du gradient conjugué, mais dans cette méthode nous avons $\textbf{d}^k=\textbf{r}^k$. De plus l'algorithme à employer est largement plus simple:

1. Calculer le residuel, $\bf r^k$, qui sert également de vecteur de direction, $\bf d^k$
2. Calculer $\alpha$
3. Calculer ${\bf p}^{k+1}={\bf p}^k + \alpha {\bf d}^k$
4. Repeter!

Ici aussi il faut forcer les bonnes conditions aux bord sur $\textbf{r}^k$ puisque :
\begin{equation}
{\bf p}^{k+1}={\bf p}^k + \alpha {\bf d}^k={\bf p}^k + \alpha {\bf r}^k
\end{equation}



Nous avons donc défini une fonction qui prend comme argument une matrice initiale p, la discrétisation selon dx, et une condition "L1_target" qui une fois vérifiée arrête le calcul, et qui renvoie la matrice p après que la coniditon soit vérifiée, ainsi qu'une liste contenant les normes L1 calculées. 

```python
def steepest_descent_2d(p, dx, l1_target):
    '''
    Calcule la solution à l'équation biharmonique dans une cavité dont l'un des 
    bords se mouvoit à vitesse constante, par la méthode de plus forte pente.
    Assume les conditions aux bords de Neumann, mais aussi qu'il n'y a pas de flôt
    en dehors de la cavité.

    Parameters:
    ----------
    p : Matrice 2D de flottants
        Matrice initiale
    dx: flottant
        Discrétisation selon x 
    l1_target: float
        critère d'intérruption d'itérations
        
    Returns:
    -------
    p: Matrice 2D de flottants
         Résultat de l'équation biharmonique.
    '''
    
    ny, nx = p.shape
    r  = numpy.zeros((ny,nx)) # residuel
    Ar  = numpy.zeros((ny,nx)) 
    
    l1_norm = 1
    iterations = 0
    l1_conv = []
    
    # Iterations
    while l1_norm > l1_target:

        pd = p.copy()
        
        #Calcul du résiduel : Step 1
        r[2:-2,2:-2]=(-p[4: ,2:-2]-p[2:-2,4: ]-p[:-4 ,2:-2]-p[2:-2,:-4 ]-2*p[3:-1,3:-1]\
                 -2*p[3:-1,1:-3]-2*p[1:-3,3:-1]-2*p[1:-3,1:-3] +8*p[3:-1,2:-2]\
                 +8*p[2:-2,3:-1]+8*p[1:-3,2:-2]+8*p[2:-2,1:-3] -20*p[2:-2,2:-2])/dx**4
        
        #Conditions aux bord implémentées sur r.
        r[0,:]=r[-1,:]=r[:,0]=r[:,-1]=0 #Conditions aux bords de Dirichlet
    
        r[-2,2:-2]=-dy/2+(1/4)*r[-3,2:-2] #Conditions aux bords de Neumann
        r[1,2:-2]=(1/4)*r[2,2:-2]         #Conditions aux bords de Neumann
        r[1:-1,1]=(1/4)*r[1:-1,2]         #Conditions aux bords de Neumann
        r[1:-1,-2]=r[1:-1,-3]*(1/4)       #Conditions aux bords de Neumann
        
        
        Ar[2:-2,2:-2]=- (-r[4:,2:-2]-r[2:-2,4:]-r[:-4,2:-2]-r[2:-2,:-4]-2*r[3:-1,3:-1]\
                 -2*r[3:-1,1:-3]-2*r[1:-3,3:-1]-2*r[1:-3,1:-3] +8*r[3:-1,2:-2]\
                 +8*r[2:-2,3:-1]+8*r[1:-3,2:-2]+8*r[2:-2,1:-3] -20*r[2:-2,2:-2])/dx**4
    
        rho = numpy.sum(r*r) #Calcul du numérateur de alpha
        sigma = numpy.sum(r*Ar)#Calcul du dénominateur de alpha
        alpha = rho/sigma  #Calcul de alpha : Step 2
        

        p = pd + alpha*r # Calcul de p : Step 3
        
        p[0,:]=p[-1,:]=p[:,0]=p[:,-1]=0 #Conditions aux bords de Dirichlet
        
        p[-2,2:-2]=-dy/2+(1/4)*p[-3,2:-2] #Conditions aux bords de Neumann
        p[1,2:-2]=(1/4)*p[2,2:-2]         #Conditions aux bords de Neumann
        p[1:-1,1]=(1/4)*p[1:-1,2]         #Conditions aux bords de Neumann
        p[1:-1,-2]=p[1:-1,-3]*(1/4)       #Conditions aux bords de Neumann
        
        
        l1_norm = L1norm(pd,p)
        iterations += 1
        l1_conv.append(l1_norm)
    
    print('Nombre d itérations par la méthode de pente la plus descendante: {0:d}'.format(iterations))
    return p, l1_conv     

```

Pour comparer les méthodes, on pose une condition de fin d'itération pas trop grande, car les résultats seront largement assez représentatifs avec cette condition.

```python
l1_target_2 = 1e-4 # Condition moins restrictive pour sortir de la boucle
p_steep, l1_conv_steep =steepest_descent_2d(p_i.copy(), dx, l1_target_2)
p_cg_2, l1_conv_cg_2 = gradient_conj(p_i.copy(), dx, l1_target_2)
```
