## Question 1

$\textit{Montrer qu'en vertu des équations de Maxwell considérées, le champ magnétique obéit à une équation de diffusion.}$

Tour d'abord, la symétrie nous permet de simplifier le problème. En effet on considère que le modèle s'étant à l'infini selon $O_{z}$, on peut donc considérer que le problème est symétrique selon l'axe $O_{z}$ et cela nous permet de travailler dans une coupe transverale du plan $O_{xy}$.


L'évolution du champ électromagnétique dans le milieu considéré est gouvernée par les équations de Maxwell: $$ \nabla \times \textbf{E}  = -\partial_{t} \textbf{B} \mbox{ (1)}\\
                                  \nabla \times \textbf{B} = \mu_{0}\textbf{J}\mbox{ (2)}\\
                                  \nabla . \textbf{B} = 0\mbox{ (3)}$$
Où le courant de déplacement a été négligé car on est dans un régime quasi-stationnaire.
En supposant que la loi d'Ohm s'applique dans tout le système, on a l'expression: $$ \frac{\textbf{J}}{\sigma} = \textbf{E} \mbox{ (4)}$$
avec $\sigma$ la conductivité qui dépend du matériau considéré, $\textbf{J}$ la densité de courant et $\textbf{E}$ le champ életrique. 
Si on prend le rotationel de l'équation (2) de chaque côté, on obtient:
$$\nabla\times(\nabla \times\textbf{B})= \mu_{0}(\nabla\times \textbf{J})$$
En prenant l'équation (4), on a:
$$\nabla\times\textbf{J}=\sigma(\nabla \times \textbf{E})= - \sigma \partial_{t}\textbf{B}$$
On a donc :
$$\nabla(\nabla.\textbf{B})-\Delta\textbf{B} = -\mu_{0}\sigma \partial_{t} \textbf{B}$$
En prenant en compte l'équation 3, on a donc finalement:
$$\Delta{\textbf{B}} = \sigma \mu_{0}(\partial_{t}\textbf{B})$$
On obtient bien une équation de diffusion du champ $\textbf{B}$:
$$\boxed{\partial_{t}\textbf{B}= \frac{1}{\sigma\mu_{0}}{\Delta}\textbf{B}}$$
avec $\frac{1}{\sigma \mu_{0}}$ le coefficient de diffusion du champ magnètique dans le conducteur.






## Question 2

$\textit{Quelles sont les conditions aux bords pour le champ magnétique dans la configuration choisie?}$

##### Bord intérieur:
Pour trouver la condition sur le bord du supraconducteur, on utilise le fait que la composante tangentielle du champ éléctrique est continue au travers de la surface. En considérant le milieu 1 et 2 comme étant le supraconducteur et le conducteur respectivement, la propriété précédente se traduit par l'expression:

$$\mathbf{n_{21}} \times (\mathbf{E_{2}}-\mathbf{E_{1}})=\mathbf{0} \mbox{ (5)}$$

Avec $\mathbf{n_{21}}$ le vecteur normal sortant du milieu 1 (et donc du supraconducteur) vers le milieu 2 (donc du conducteur). On remarque que ce vecteur sera toujours placé dans le plan $O_{xy}$ à cause de la symétrie du problème. Etant donné que la conductivité $\sigma$ du supraconducteur est infinie, la limite lorsque $\sigma$ tend vers l'infini de l'expression (4) nous permet de déduire que le champ électrique dans ce milieu est nul,c'est-à-dire $\mathbf{E_{1}}=\mathbf{0}$. L'équation de Maxwell (1) nous permet de déduire que le champ magnétique à l'intérieur du supraconducteur est constant:

$$-\partial_{t} \mathbf{B_{1}}=\mathbf{0} \Rightarrow \mathbf{B_{1}}=cst$$

Plus précisément, il est nul car initialement, le courant électrique et le champ éléctromagnétique sont égaux à zéro partout dans le domaine : $\mathbf{B_{1}}=\mathbf{0}$

L'expression (5) devient donc:
$$\mathbf{n_{21}} \times \mathbf{E_{2}}=\mathbf{0}$$
De plus, le champ électrique dans le milieu 2 peut s'exprimer par: 
$$ \mathbf{E_{2}} = \frac{\textbf{J}}{\sigma} = \frac{\nabla \times \textbf{B}}{\mu_{0}\sigma}$$ 

D'autre part, grâce à la symétrie du problème et grâce à la loi de Biot et Savart

$$\mathbf{B}(\mathbf{r}) = \frac{\mu_{0}}{4\pi} \oint_C\frac{I\mathbf{dl} \times \mathbf{(r-r')}}{|r-r'|^3}$$

où C est une courbe géométrique représentant le circuit filiforme, $\mathbf{r'}$ est un point de cette courbe, $\mathbf{dl}$ est le vecteur déplacement élémentaire tangent à la courbe C au point $\mathbf{r'}$. Ce circuit parcouru par un courant d'intensité I continu  crée, en tout point $\mathbf{r}$ extérieur à C, le champ magnétique $\mathbf{B}$($\mathbf{r}$) donné par l'expression ci-dessus. Cette formule reste tout à fait analogue à une formule pour un courant de surface J. Ici ce qui nous intéresse est le produit vectoriel dans cette formule, puisqu'il va nous permettre de déterminer le sens du champ magnétique dans le conducteur.

Le champ magnétique dans le conducteur est donc de la forme $\textbf{B}=(0,0,-B)$. 

On a de plus:
$$\frac{\nabla \times \textbf{B}}{\mu_{0}}= \frac{1}{\mu_{0}}(-\partial_{y}B \mathbf{u_{x}}+ \partial_{x} B \mathbf{u_{y})}$$
Et donc:
$$\mathbf{n_{21}}\times (-\partial_{y}B \mathbf{u_{x}} + \partial_{x} B \mathbf{u_{y}})=0 \mbox{ (6)}$$

A partir de là, on va considérer les 4 bords du supraconduteur séparément. On s'intéresse tout d'abord au bord supérieur, parallèle à $O_{x}$. Dans ce cas là, la normal $\mathbf{n_{21}}$ est de la forme $(0,1,0)$. Dans ce cas là, le produit vectoriel (6) nous donne:

$$\boxed{-\partial_{y} B \mathbf{u_{z}}=0}$$

Ensuite, on considère le bord inférieur, parallèle à $O_{x}$. Dans ce cas là: $\mathbf{n_{21}}=(0,-1,0)$. Par la même méthode que précédemment, on obtient:

$$\boxed{\partial_{y} B \mathbf{u_{z}}=0}$$

Les deux derniers bords se calculent de la même manière pour $\mathbf{n_{21}}=(1,0,0)$ (bord de droite, parallèle à $O_{y}$) et $\mathbf{n_{21}}=(-1,0,0)$ (bord de gauche, parralèle à $O_{y}$) et on obtient respectivement les conditions:

$$\boxed{\partial_{x} B \mathbf{u_{z}}=0}$$

$$\boxed{-\partial_{x} B \mathbf{u_{z}}=0}$$

Qui sont en fait les mêmes expressions.


Nos conditions aux 4 bords du supraconducteur s'expriment sous forme de dérivées partielles. Plus précisement, nos conditions de bords spécifient la dérivée de valeur de la solution du champ magnétique $\mathbf{B}$ aux 4 bords du supraconducteur. Ce sont donc des conditions aux bords de Neumann.

Pour exprimer numériquement ces 4 conditions, on peut réécrire plus précisément:


\begin{equation}
\left. {-\partial_{y} B}\mathbf{u_{z}}\right|_{y=1cm} = 0
\end{equation}

\begin{equation}
\left. {\partial_{x} B}\mathbf{u_{z}}\right|_{x=1cm} = 0
\end{equation}

\begin{equation}
\left. {\partial_{y} B}\mathbf{u_{z}}\right|_{y=0.5cm} = 0
\end{equation}

\begin{equation}
\left. {-\partial_{x} B}\mathbf{u_{z}}\right|_{x=0.5cm} = 0
\end{equation}

Ici on a pris les positions du supraconducteur énoncées dans le problème, mais nous avons codé de telle manière à ce que les conditions aux bords soient encore appliquables pour un problème où les longueurs du supraconducteur ne sont pas celles de l'énoncé.

Physiquement, il y a discontinuité entre le conducteur et le supraconducteur mais on ne souhaite pas que cette discontinuité influence la diffusion du champ magnétique dans le conducteur. On impose donc que la valeur du champ magnétique ne change pas dans les directions $x$ et $y$ près des bords extérieurs du supraconducteur. C'est ce que les 4 équations précédentes expriment exactement.


#### Bord extérieur:

Pour connaitre la valeur du champ magnétique au bord extérieur du conducteur, on utilise la loi d'Ampère. On a:
$$\int_{S} \nabla \times \mathbf{B}.\mathbf{dS}= \int_{S} \mu_{0} \mathbf{J}.\mathbf{dS}$$
pour chacuns des 4 bords externes du conducteur.

On va regarder les deux termes de cette équation séparément.

Regardons d'abord le terme de gauche, par le théorème de Stokes on peut écrire:
$$\int_{S} \nabla \times \mathbf{B}.\mathbf{dS}= \oint_{\partial S} \mathbf{B}.\mathbf{dl} \mbox{ (7)}$$

$S$ sont des surfaces formées par des cadres parallèles au plan $O_{xz}$ qui encardrent les bords externes du conducteur. $\partial S$ est le contour de cette boucle et $\mathbf{dl}$ est orienté en suivant le cadre. Par exemple, en prenant le bord parralèle à $O_{y}$ en $x=0$, le $\mathbf{dl}$ du bord inférieur (supérieur) du cadre est dirigé vers les $O_{x}$ positifs (négatifs) et le $\mathbf{dl}$ du bord de droite (gauche) du cadre est dirigé vers les $O_{z}$ négatifs (positifs). On se rapelle que le champ magnétique est orienté vers les $O_{z}$ négatifs. 

Initialement, le champ est nul dans le conducteur. La valeur de $\mathbf{B}.\mathbf{dl}$ est donc nulle sur le bord droit du cadre. De plus, les comportements des champs magnétiques sur les bords supérieurs et inférieurs du cadre s'annulent car dans ces cas là, les champs magnétiques sont perpendiculaires aux dl respectifs. L'expression (7) ne prend donc en compte que le bord gauche du cadre. On a donc:

$$\oint_{\partial S} \mathbf{B}.\mathbf{dl}= \oint_{\partial S_{1}} -B dl = -BL$$

avec $L$ est la longueur du conducteur selon z.


Regardons maintenant le terme de droite. Sur ce bord, on remarque déjà assez facilement que la densite de courant $\textbf{J}$ n'a pas de dépendance en $y$. On peut donc écrire: 

$$\mathbf{J}=\mathbf{J}(x,z)$$
On peut écrire le terme de droite comme 
$$\int_{S} \mu_{0} \mathbf{J}.\mathbf{dS} = \int_{0}^{L}\int_{0}^{L_x} \mu_{0} J(x,z) dx dz$$
étant donné que $\mathbf{dS}$ et $\mathbf{J}(x,z)$ sont orientés de la même manière.

Cependant, initialement, la densité de courant, sur le bord extérieur inférieur du conducteur est telle que :
$$J(x,z)=1 \text{ si x=0}$$
$$J(x,z) = 0 \text{ sinon}$$

$J(x,z)$ se comporte donc comme une distribution de Dirac:
$$\int_{S} \mu_{0} \mathbf{J}.\mathbf{dS}=\int_{0}^{L}\int_{0}^{L_x} \mu_{0} J(x,z) \delta(x) dx dz = \int_{0}^{L} \mu_{0} \overset{\sim}{J}(x=0,z)dz = \mu_{0} J L$$

avec $$\delta(x)=1 \text{ si x=0}$$
$$\delta(x) = 0 \text{ sinon}$$

En prenant $J= 1A/m^{-2}$ et on égalisant ls deux résultats:
$$\boxed{-B = \mu_{0}}$$

Physiquement, le champ magnétique sur le bord externe du conducteur est constant tout au long de la diffusion et vaut $-\mu_{0}= -1,26.10^{-6} N/A^2$. Notre condition de bord spécifient la valeur de la solution du champ magnétique $\mathbf{B}$ aux 4 bords externes du conducteurs, c'est donc une conditon de Dirichlet. 








## Question 3

$ \textit{Calculer l'évolution au cours du temps du champ magnétique par la méthode d'Euler (forward) dans le temps, et avec une discrétisation centrée d'ordre 2 pour le laplacien.} $

Commençons d'abord par importrer les librairies et fonctions dont nous avons besoin, ici Numpy et Matplotlib. De Matplotlib nous importons pyplot qui va nous permettre de générer des graphiques. Nous prenons la même police de caligraphie que dans le mooc.

```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams, cm
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

Nous définissions d'abord les conditions initiales du problème, ainsi que quelques nouvelles "variables" que l'on pourra ajuster en fonction du problème étudié. Nous avons supposé, que le problème possèdera cependant la même symétrie, c'est à dire, le supraconducteur sera toujours contenu dans le conducteur, et il sera toujours au centre de ce conducteur. Cependant, rien n'empêche que le conducteur soit rectangulaire, et le supraconducteur carré par exemple. Nous avons également supposé que le conducteur et le supraconducteur ont des bords droits.

L'enoncé nous demande deux choses:

$\textit{Prendre 121 points dans chaque direction $(x,y)$ pour discrétiser une section du domaine: $nx \times ny =121\times 121$.}$

$\textit{Prendre le pas de temps comme $dt=2.5\cdot 10^{-7}s$ (justifier ce choix).}$

Donc imposons les, dès le départ. $\textbf{Lx}$, et $\textbf{Ly}$ sont les longueurs du conducteur selon x et y respectivement, et $\textbf{cond_x}$ et $\textbf{cond_y}$ le longueurs du supraconducteur, selon x et y respectivement.


```python
nx=121                             # Nombre de colonnes dans la matrice.
ny=121                             # Nombre de lignes dans la matrice.

Lx=0.015                           # Longueur dans la direction x du conducteur.
Ly=0.015                           # Longueur dans la direction y du conducteur.

cond_x=0.005                       # Longueur du supraconducteur dans la direction x.
cond_y=0.005                       # Longueur du supraconducteur dans la direction y.

x=numpy.linspace(0,Lx,nx)          # Liste de nx éléments, espacés entre-eux linéairement, dans la direction x.
y=numpy.linspace(0,Ly,ny)          # Liste de ny éléments, espacés entre-eux linéairement, dans la direction y. 

dx=Lx/(nx-1)                       # Pas dx dans la direction x.
dy=Ly/(ny-1)                       # Pas dy dans la direction y.

sigma = 5.96*10**7                 # Donnée du problème.
mu_0=1.26*10**(-6)                 # Donnée du problème. 
alpha = 1/(mu_0*sigma)             # Utilisé plus tard pour simplifier les calculs.

dt = 2.5*10**(-7)                  # Pas de temps dt demandé dans l'énoncé.
```

Ici on définit plusieures variables, qui vont nous permettre d'ajuster la fonction qui calculera B en fonction du problème posé. Pour ce problème nous allons créer une matrice B, qui va contenir les valeurs du champ B, dans l'espace considéré. Cette matrice correspondra aux différents endroits où le champ magnétique va être calculé. Par exemple $B[0,0]$ correspond au champ magnétique en position $x=0$  et $y=0$. N'oublions pas que pour une matrice $B[j,i]$, l'indice j correspondra à la position par rapport à l'axe y et l'indice i, lui, correspondra à la position par rapport à l'axe des x. Plus explicitement, $\textbf{rap_x}$ définit le rapport entre la longueur du conducteur et la longueur du supraconducteur selon x, $\textbf{deb_x_supra}$ définit l'indice dans la matrice B de la position du bord gauche du supraconducteur, et $\textbf{fin_x_supra}$ l'indice dans la matrice B de la position du bord droit du supraconducteur. De manière analogue, $\textbf{rap_y}$ définit le rapport entre la longueur du conducteur et la longueur du supraconducteur selon y, $\textbf{deb_y_supra}$ l'indice dans la matrice B de la position du bord inférieur du supraconducteur et $\textbf{fin_y_supra}$ l'indice dans la matrice B de la position du bord supérieur du supraconducteur.

```python
rap_x=int(Lx/cond_x)               # Rapport entre la longueur, dans la direction x, du conducteur et du supraconducteur.
rap_y=int(Ly/cond_y)               # Rapport entre la longueur, dans la direction y, du conducteur et du supraconducteur.

deb_x_supra=int((nx-1)/rap_x)      # Elément colonne correspondant au bord gauche du supraconducteur.
deb_y_supra=int((ny-1)/rap_y)      # Elément ligne correspondant au bord inférieur du supraconducteur.

fin_x_supra=(nx-1)-deb_x_supra     # Elément colonne correspondant au bord droit du supraconducteur.
fin_y_supra=(ny-1)-deb_y_supra     # Elément ligne correspondant au bord superieur du supraconducteur.
```

Justifions, tout d'abord, le choix de dt imposé dans l'énoncé. Nous avons vu dans le chapitre 2, le nombre CFL. Ici, nous sommes demandés de calculer le champ magnétique avec la méthode d'Euler forward dans le temps, et pour cette méthode nous savons que le nombre CFL doit être au plus égal à $\frac{1}{2}$. Rappellons la définition de ce nombre, pour une équation de diffusion du type $\frac{\partial u}{\partial t}=c(\frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2})$, où $c$ est le coefficient de diffusion.Pour ce type d'équation le nombre CFL vaut:
$$ \sigma_{cfl} =\frac{c\Delta t}{\Delta x^2}+\frac{c\Delta t}{\Delta y^2} \leq \frac{1}{2}$$
où $c$ est exprimé en $m/s^2$. Ainsi nous avons ici l'équation de diffusion pour le champ magnétique (que nous avons montré un peu plus loin): $$ \frac{\partial \textbf{B}}{\partial t}= \alpha \Delta \textbf{B}$$, où $ \alpha= \frac{1}{\mu_0 \sigma}$. En prenant comme dans l'énoncé $dt= 2.5 \times 10^{-7}s$ , nous obtenons $\sigma_{cfl}$ suivant:

```python
sigma_cfl= 1/(mu_0*sigma)*(dt/dx**2+dt/dy**2)
print(sigma_cfl)
```

Les solutions obtenues seront donc bien stables.

(Source: https://en.wikipedia.org/wiki/Courant%E2%80%93Friedrichs%E2%80%93Lewy_condition)

Nous allons maintenant construire notre fonction qui calculera le champ magnétique diffusé dans le conducteur, en prenant bien soin d'imposer les conditions aux bords. Nous appelons cette fonction $fcts$, elle prend comme argument une matrice B de départ, un temps nt, qui constitue le nombre d'itérations à faire pour avoir accès aux valeurs du champ magnétique à un temps t car $t=nt\times dt$, dt le pas de temps choisi, dx et dy les pas correspondant à la distance entre chaque point où l'on va calculer le champ magnétique.

Nous devons utilisé la méthode D'Euler, avec une discrétisation centrée d'ordre deux pour le Laplacien. Rappelons que l'équation de diffusion du champ magnétique, avec $\alpha$ défini ci-dessus, s'écrit : $$ \frac{\partial \textbf{B}}{\partial t} =  \alpha \Delta \textbf{B} $$

Si nous discrétisons cette équation nous obtenons :

$$ \frac{B_{i,j}^{n+1} - B_{i,j}^n}{\Delta t}=\alpha (\frac{B_{i+1,j}^n - 2B_{i,j}^n + B_{i-1,j}^n}{\Delta x^2}+\frac{B_{i,j+1}^n - 2B_{i,j}^n + B_{i,j-1}^n}{\Delta y^2})$$ où les indices n correspondent aux itérations sur le temps, les indices i à la position selon x et les indices j à la position selon y. En mettant le terme $B_{i,j}^{n+1}$ en évidence, nous obtenons:$$B_{i,j}^{n+1} =B_{i,j}^n + \alpha (\frac{\Delta t}{\Delta x^2}(B_{i+1,j}^n - 2B_{i,j}^n + B_{i-1,j}^n)+\frac{\Delta  t}{\Delta y^2}(B_{i,j+1}^n - 2B_{i,j}^n + B_{i,j-1}^n)) \text{    (1)}$$
C'est cette équation que nous allons utilisé pour résoudre le problème numériquement. 

Nous devons également imposer les conditions aux bords à tout instant. Nous avons la condition de Dirichlet $$ B=-\mu_0$$ sur tous les bords extérieurs du conducteur. Nous devons également imposer, à tout instant, que le supraconducteur possède un champ magnétique nul. Nous devons également imposer que le supraconducteur n'influence pas la diffusion du champ magnétique dans le conducteur. En d'autres termes, nous devons imposer que le gradient du champ magnétique soit nul aux bords du supraconducteur. Explicitement: $\nabla \textbf{B} = \mathbf{0}$ entre le conducteur et le supraconducteur. Ainsi on à une condition de Neumann à imposer aux bords du supraconducteur. Nous imposons ici cette condition de bord sur les points évalués juste avant le supraconducteur, car il doit y avoir une discontinuité entre le supraconducteur et le conducteur.
    

 Définissons notre fonction qui va calculer le champ magnétique B. Cette fonction prend en argument une matrice B, correspondant aux conditions initiales (avant la diffusion) du champ B, nt le nombre d'itérations à faire pour avoir accès à l'instant t auquel on souhaite calculer le champ B et alpha, défini plus haut. Cette fonction va aussi prendre comme arguments dt, dx, dy, les pas de temps et de positions respectivement. De plus les deux derniers arguments (Tot, r) seront chacun une liste de taille $\frac{nt}{r}\times nx \times ny$, nommée Tot, où l'on va stocker les valeurs du champ magnétique toutes les $r$ itérations, r étant un paramètre arbitaire. Cette matrice nous sera utile pour l'animation. 

```python
def ftcs(B, nt, alpha, dt, dx, dy, Tot, r):
    
    for n in range(nt):
        Bn = B.copy()                   # On crée une copie à chaque itération, puisqu'on va manipuler B à chaque itération.
 
        B[-1,:] = B[:,-1] = B[0,:]=B[:,0] = -mu_0           # Condition de Dirichlet sur le bord du conducteur extérieur.
        
        B[1:-1,1:-1] = Bn[1:-1,1:-1] + alpha *\
            (dt/dy**2 * (Bn[2:,1:-1] - 2*Bn[1:-1,1:-1] + Bn[:-2,1:-1]) +\
             dt/dx**2 * (Bn[1:-1,2:] - 2*Bn[1:-1,1:-1] + Bn[1:-1,:-2]))          # Implémentation de l'équation 1.
            
        B[deb_y_supra:fin_y_supra+1,deb_x_supra-1]=B[deb_y_supra:fin_y_supra+1,deb_x_supra-2] #Condition de Neumann sur le bord gauche du supraconducteur.
        B[deb_y_supra:fin_y_supra+1,fin_x_supra+1]=B[deb_y_supra:fin_y_supra+1,fin_x_supra+2] #Condition de Neumann sur le bord droit du supraconducteur.
        B[deb_y_supra-1,deb_x_supra:fin_x_supra+1]=B[deb_y_supra-2,deb_x_supra:fin_x_supra+1] #Condition de Neumann sur le bord inférieur du supraconducteur.
        B[fin_y_supra+1,deb_x_supra:fin_x_supra+1]=B[fin_y_supra+2,deb_x_supra:fin_x_supra+1] #Condition de Neumann sur le bord supérieur du supraconducteur.
        
        B[deb_y_supra:fin_y_supra+1,deb_x_supra:fin_x_supra+1]=0   # Condition sur le champ magnétique dans le supraconducteur
        if (n/r==int(n/r)) :
             
            Tot[int(n/r)]=B        # Matrice où on stocke les valeurs de B une itération sur r.
    return B
```

Maintenant que nous avons notre fonction qui calcule le champ B à un temps nt, calculons ce champ à un certain temps t, arbitrairement choisi. N'oublions pas que $t=nt \times dt$. Pour $t=2.5 \times 10^{-4}s$ on a $nt=1000$.

```python
Bi_test = numpy.zeros((ny, nx))     # Le champ B est nul partout au départ.
nt=1000                             # On calcule le champ B après un temps t=2.5*10*(-4)s, correspondant à nt=1000.
r=100                               # On veut remplir T par les valeurs de B une fois toutes les 100 itérations.
Tot=numpy.zeros((int(nt/r),nx,ny))  # Matrice qui va contenir B calculé à chaque itération.
```

```python
B1_test = ftcs(Bi_test.copy(),nt, alpha, dt, dx, dy,Tot,r)  
# Calcul du champ B à l'instant nt, avec un pas dt en temps, et un pas de temps dx,dy en espace.
```

Créeons un graphique de ce champ magnétique à l'instant t.

```python
pyplot.figure(figsize=(7,5))
pyplot.contourf(x,y,B1_test,150,cmap=cm.viridis)  # On choisi 150 couleurs sur notre graphique.
pyplot.xlabel('$x$')
pyplot.ylabel('$y$')
pyplot.colorbar();
```

On obtient bien un champ magnétique nul dans le supraconducteur ainsi qu'un champ magnétique négatif dans le conducteur à cause de la direction du courant $\textbf{J}$ appliqué. On voit aussi très clairement la diffusion du champ magnétique, les bords correspondent aux valeurs du champ magnétique le plus grand, et plus on se rapproche du supraconducteur plus le champ magnétique est faible.

Remarquons que si l'on prend un nt suffisament grand, c'est à dire si on attend suffisament longtemps, le champ magnétique dans le conducteur va s'uniformiser. Regardons pour $t=2.5\times 10^{-3}s$, donc $nt=10000$:

Pour calculer $\textbf{J}$, nous devons utiliser une des équations de Maxwell : $$\nabla \times \textbf{B} = \mu_0 \textbf{J}$$
Rappelons que le rotationnel s'écrit: $$\nabla \times \textbf{B} =  (\frac{\partial B_z}{ \partial y} - \frac{\partial B_y}{\partial z},\frac{\partial B_x}{ \partial z} - \frac{\partial B_z}{\partial x},\frac{\partial B_y}{ \partial x} - \frac{\partial B_x}{\partial y})=(J_x,J_y,J_z)\mu_0$$

Or nous savons que les composantes $B_x$ et $B_y$ sont nulles, puisque le champ est seulement dans la direction z. Ainsi nous avons:
$$\textbf{J}=\frac{1}{\mu_0}(\frac{\partial B_z}{\partial y},-\frac{\partial B_z}{\partial x},0)$$
En prenant une discrétisation d'ordre 1, nous avons:
$$J_x=\frac{B_z^{x,y+1}-B_z^{x,y}}{\mu_0\Delta y}$$
$$J_y=\frac{-B_z^{x+1,y}+B_z^{x,y}}{\mu_0\Delta y}$$
$$J_z=0$$

Nous allons utiliser ces trois équations pour trouver $\textbf{J}$.

```python
nt2=int(t2/dt)        # Nombre d'itérations que l'on doit faire pour avoir accès à la valeur du champ B à l'instant t2 demandé dans l'énoncé.

m2=int(x2/dx)         # Indice correspondant à la position x pour laquelle on veut calculer le courant J.
n2=int(y2/dy)         # Indice correspondant à la position y pour laquelle on veut calculer le courant J.

Bi_J = numpy.zeros((ny, nx))                         # Le champ B est initialement nul partout.
r=10                                                 # Nous voulons voir la diffusion du champ magnétique jusqu'au temps 
                                                     # demandé dans l'énoncé donc nous choissons r =10.
Tot=numpy.zeros((int(nt2/r),nx,ny))
                
B_J = ftcs(Bi_J.copy(),nt2, alpha, dt, dx, dy,Tot,r)   # On calcule le champ B après un diffusion pendant un temps t2.

Jx= (B_J[n2+1,m2]-B_J[n2,m2])/(dy*(mu_0))            # Implémentation de la discrétisation du rotationnel pour trouver 
Jy= -(B_J[n2,m2+1]-B_J[n2,m2])/(dx*(mu_0))           # les composantes de J.
Jz=0

print('Le vecteur J en t2 vaut (',Jx,',',Jy,',',Jz,') Ampères par mètres carrés.')
```

## Question 7

$\textit{Que vaut, à l'instant $t=1.25 \cdot 10^{-2}s$, la valeur du flux du champ magnétique à  travers} $
$\textit{ la section $z=0$ du conducteur?}$

Etablissons les conditions initiales du problème.

```python
t3= 1.25*10**(-2)                               # Temps auquel on doit évaluer B.
nt3= int(t3/dt)                                 # Nombre d'itérations que l'on doit faire pour avoir accès au valeurs du champ B
                                                # au temps t3 demandé dans l'énoncé.

Bi_flux = numpy.zeros((ny, nx))                      # Le champ magnétique est initialement nul.
Tot=numpy.zeros((nt3,nx,ny))
r=nt3
B_flux= ftcs(Bi_flux.copy(), nt3, alpha, dt, dx, dy,Tot,r)      # On calcule le champ magnétique après diffusion pendant un temps t3.
```

Pour calculer le flux du champ magnétique, nous devrions calculer $$\int_0^{Lx}\int_0^{Ly} \textbf{B.dS}$$ où $\textbf{dS}$ est le vecteur normal à un élement infinitésimal de surface. Cependant ce genre de calcul n'est pas possible sur ordinateur, il faut donc discrétiser, avec une somme. On a donc:$$\Sigma_{i} \textbf{B}_i\textbf{.dS}_i$$
En prenant des $\textbf{dS}_i$ qui sont tous égaux, ç-à-d, $\textbf{dS}_i=\textbf{n}dxdy$ où $\textbf{n}$ est le vecteur normal à l'élément de surface $dxdy$, alors on peut le mettre en facteur dans la somme. Ainsi on a donc:$$dxdy\Sigma_{i}B_{i}$$ Cependant nous ne connaissons que les valeurs de B à certains endroits, à cause de la discrétisation. Nous choissons donc des surfaces $dxdy$ où notre valeur de champ B sera au centre de ces surfaces. Ceci équivaut à "décaler" les surfaces de $\frac{dx}{2}$ et de $\frac{dy}{2}$ sans compter les surfaces qui ne sont pas entièrement contenues dans le conducteur. 

Nous perdons alors une surface proche du bord, qui est de taille $((nx-2)\times dx \times \frac{dy}{2})*2 +((ny-2)\times dy \times \frac{dx}{2})*2+ 4*\frac{dxdy}{4}$. Le premier terme correspond aux deux bandes sur les cotés gauche et droit du conducteur, le deuxième terme aux deux bandes supérieure et inférieure du conducteur, et le dernier aux 4 coins reliant ces bandes. Cependant, dans la limite ou $dx$ et $dy$ tendent vers 0, cette surface est tout à fait négligéable par rapport au reste du problème.

Nous choisissons cependant de compter les surfaces de bord restantes, car ici la discrétisation est grande, et le code n'est pas très difficile. Pour cela nous utilisons une moyenne pondérée.Commençons par les bandes, par exemple la bande gauche du conducteur. Pour celle-ci, un petit calcul de géométrie élémentaire nous permet de savoir que le champ B en son centre est représenté par la moyenne pondérée suivante:$\frac{3}{4}B[j,0] +\frac{1}{4}B[j,1]$, où j parcours les valeurs de 1 à (ny-2). De manière analogue pour les 3 bandes restantes, nous allons alors avoir le champ B qui vaut:
$$\frac{3}{4}B[j,0] +\frac{1}{4}B[j,1], \text{pour le bord gauche du conducteur}$$
$$\frac{3}{4}B[j,nx-1] +\frac{1}{4}B[j,nx-2], \text{pour le bord droit du conducteur}$$
$$\frac{3}{4}B[0,0] +\frac{1}{4}B[1,i], \text{pour le bord inférieur du conducteur}$$
$$\frac{3}{4}B[ny-1,i] +\frac{1}{4}B[ny-2,i], \text{pour le bord supérieur du conducteur}$$
où i et j parcourent les valeurs 1 à (nx-2) et 1 à (ny-2) respectivement. Pour ces 4 bandes, le flux sera donné par:$B_pondéré*\frac{dxdy}{2}$. Il suffit maintenant de compter les 4 coins restants.

Pour ceux-ci le théorème de pythagore nous assure que nous pouvons implémenter la même moyenne pondérée, pour trouver le champ magnétique au milieu de cette surface de taille $\frac{dxdy}{4}$. Ainsi nous avons, par exemple pour le champ magnétique dans le coin inférieur gauche, la pondération: $B= \frac{3}{4}B[0,0]+ \frac{1}{4}B[1,1]$
Nous aurons alors:
$$B= \frac{3}{4}B[0,0]+ \frac{1}{4}B[1,1], \text{pour le coin inférieur gauche}$$
$$B= \frac{3}{4}B[0,nx-1]+ \frac{1}{4}B[1,nx-2], \text{pour le coin inférieur droit}$$
$$B= \frac{3}{4}B[ny-1,0]+ \frac{1}{4}B[ny-2,1], \text{pour le coin supérieur gauche}$$
$$B= \frac{3}{4}B[ny-1,nx-1]+ \frac{1}{4}B[ny-2,nx-2], \text{pour le coin inférieur droit}$$


Nous allons créer une fonction qui calcule donc le flux dans cette surface. Notre fonction, calcul_flux, prend comme variables, une matrice B_flux contenant les valeur du champ magnétique à un instant t, nx, ny le nombre de points dans cette matrice et dx, dy les longueurs des surfaces discrétisées. La fonction va ensuite sommer tous ces champs magnétiques et nous renvoyer le flux du champ magnétique selon la formule:
$$ \phi = Flux_{int}*dxdy+Flux_{bandegauche}*\frac{dxdy}{2}+Flux_{bande droite}\frac{dxdy}{2}+Flux_{bande supérieure}*\frac{dxdy}{2}+Flux_{bandeinférieure}*\frac{dxdy}{2}+Flux_{4coins}*dx*dy$$

```python
def calcul_flux(B_flux,nx,ny,dx,dy):
    NormB_inf=0            # NormB_inf correspond à la somme des champs magnétiques sur la bande inférieure.
    NormB_sup=0            # NormB_sup correspond à la somme des champs magnétique sur la bande supérieure.
    NormB_gauche=0         # NormB_gauche correspond à la somme des champs magnétique sur la bande gauche.
    NormB_droite=0         # NormB_droite correspond à la somme des champs magnétique sur la bande droite.
    NormB_int=0            # NormB va contenir la somme de tous les champs magnétiques contenus dans le conducteur/supraconducteur.
    for j in range(1,ny):    # On va sommer sur tous les éléments de la matrice B en ne comptant pas les terme aux bords, donc
        for i in range(1,nx):  # on commence au terme B[1,1].
            NormB_int=NormB_int + B_flux[j,i] 
    for l in range(1,ny-1):
        NormB_gauche=NormB_gauche+(3/4)*B_flux[l,0]+(1/4)*B_flux[l,1]  # Calcul de la somme des champs magnétique de la bande gauche.
        NormB_droite=NormB_droite+(3/4)*B_flux[l,nx-1]+(1/4)*B_flux[l,nx-2] # Calcul de la somme des champs magnétique de la bande droite.
    for d in range(1,nx-1):
        NormB_inf=NormB_inf+(3/4)*B_flux[0,i]+(1/4)*B_flux[1,i] # Calcul de la somme des champs magnétique de la bande inférieur.
        NormB_sup=NormB_sup+(3/4)*B_flux[ny-1,i]+(1/4)*B_flux[ny-2,i] # Calcul de la somme des champs magnétique de la bande supérieure.
        
        
    B_coin1=B_flux[0,0]*(3/4)+B_flux[1,1]*(1/4)              # Calcul du champ magnétique au coin inférieur gauche.
    B_coin2=B_flux[0,nx-1]*(3/4)+B_flux[1,nx-2]*(1/4)        # Calcul du champ magnétique au coin inférieur droit.  
    B_coin3=B_flux[ny-1,0]*(3/4)+B_flux[ny-2,1]*(1/4)        # Calcul du champ magnétique au coin supérieur gauche
    B_coin4=B_flux[ny-1,nx-1]*(3/4)+B_flux[ny-2,nx-2]*(1/4)  # Calcul du champ magnétique au coin supérieur droit.
    Norm_coin = B_coin1 + B_coin2 + B_coin3 + B_coin4
    
    flux = NormB_int*dx*dy+ Norm_coin*dx*dy +(dx*dy/2)*(NormB_gauche + NormB_droite + NormB_inf + NormB_sup) 
    # Calcul du flux par la formule ci-dessus.
    return flux
```

```python
print('La valeur du flux du champ magnétique est',calcul_flux(B_flux,nx,ny,dx,dy),'Tesla mètres carrés.')
```

Lorsque nous n'avions pas tenu compte des bandes de surfaces manquantes, nous obtenions, grossièrement -2,5040 Tesla mètres carrés, la différence n'est donc pas négligeable dans notre cas, puisque la différence se fait sentir au deuxième chiffre après la virgule.
